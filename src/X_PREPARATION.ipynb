## IMF

```python
imf = pd.read_csv("../data/WEOApr2022all.csv", sep=";")[:-2]
imf
```

```python
imf["Subject Descriptor"].unique()
```

```python
# melt year rows 1980-2027 to one year column

years = list()
for y in range(1980, 2028):
    years.append(str(y))
print(years)

imf = imf.melt(id_vars=["ISO", "Country", "WEO Country Code", "WEO Subject Code", "Subject Descriptor", "Subject Notes", "Units", "Scale"], var_name="YEAR", value_vars=years)
imf
```

```python
imf_indicators = imf[["WEO Subject Code", "Subject Descriptor", "Subject Notes", "Units", "Scale"]].drop_duplicates()
imf_indicators.to_csv("../data/imf_indicators.csv", index=True)
imf_indicators
```

```python
imf = imf.pivot(index=["ISO", "Country", "WEO Country Code", "YEAR"], columns=["WEO Subject Code"], values="value")
imf = imf.reset_index()
# clean: delete NaN countries and NaN column
imf = imf[imf['Country'].notna()]
# add IMF prefix
imf = imf.add_prefix('IMF_')

imf = imf.rename({"IMF_ISO": "GID_0", "IMF_Country": "COUNTRY", "IMF_YEAR":"DATE"}, axis="columns")
imf.DATE = pd.to_numeric(imf.DATE)
```

```python
imf_stat_vars = pd.read_excel("../data/Exog Vars Stationarity Selection.xlsx")
imf_stat_vars = imf_stat_vars[imf_stat_vars["Selected"] == "no"] 
imf_stat_vars["WEO Subject Code"] = "IMF_" + imf_stat_vars["WEO Subject Code"].astype(str)
imf_stat_vars
```

```python
### Delete non-stationary variables
imf = imf.drop(imf_stat_vars["WEO Subject Code"].tolist(), axis="columns")
imf
```

```python
# pivot df to create column for every country fatality time series
piv = pd.pivot_table(imf_wb, columns=["GID_0"], index=["DATE"])
piv.columns = piv.columns.map('{0[0]}_{0[1]}'.format)
piv.tail()
```

### Linear Interpolation

```python
print(piv.isna().sum().sum())
interpol = piv.interpolate(method="linear", axis=0, limit_direction="both")
print(interpol.isna().sum().sum())
print(interpol.shape)
interpol.head()
```

```python
interpol = interpol.reset_index()
interpol.head()
```

```python
import datetime as dt
# returns datetime objects
interpol.DATE = interpol.DATE.apply(lambda row: dt.datetime(row, 1, 1))
interpol = interpol.set_index(pd.DatetimeIndex(interpol.DATE), drop=True)
interpol = interpol.drop(["DATE"], axis=1)
interpol.head()
```

```python
# generate monthly data by replicating each year 12 times with correct month as index
interpol = interpol.resample("MS").ffill()
interpol
```

```python
interpol.to_csv("../data/TB013_SOCIO_ECONOMIC_INDICATORS_INTERPOL.csv", index=True)
```
