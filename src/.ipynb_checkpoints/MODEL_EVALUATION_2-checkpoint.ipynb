### TADDA-Metric from VIEWS Conflict Prediction Competition (Uppsala University)
- "MAE with sign-error penalty"

```python
def abs_error(y_true, y_pred):
    return np.abs(y_true - y_pred)
```

```python
def squared_error(y_true, y_pred):
    return (y_true - y_pred)**2
```

$$
TADDA^{(M)} = \frac{\sum_{i=1}^N |{y_{\Delta, i} - f_{\Delta, i}}|+ |f_{\Delta, i}| \times I[y_{\Delta, i}^{(\pm)} \neq f_{\Delta, i}^{(\pm)}]
\times I[ |y_{\Delta, i} - f_{\Delta, i}| > \epsilon ]
}
{N}>
$$

```python
def tadda_error(y_true, y_pred, threshold = 0.048):
    
    mae_part = abs_error(y_true, y_pred)
        
    wrong_signs = (np.sign(y_pred) * np.sign(y_true) == -1).astype('int') # 0 if correct sign; 1 if wrong sign 
    significant_diff = (mae_part > threshold).astype('int')
    
    # mae in case wrong sign and over threshold
    sign_loss = (np.abs(y_pred) * wrong_signs * significant_diff)
    
    # print("---------------------------------------------------")
    # print(mae_part)
    # print("############")
    # # print(np.sign(y_pred))
    # # print(np.sign(y_true))
    # # print(wrong_signs)
    # # print("############")
    # print(significant_diff)
    # print("############")
    # print(sign_loss)
    # print("---------------------------------------------------")


    return mae_part + sign_loss
```

```python
def tadda_error(y_true, y_pred, threshold = 0.048):
    
    mae_part = abs_error(y_true, y_pred)
        
    wrong_signs = (np.sign(y_pred) * np.sign(y_true) == -1).astype('int') # 0 if correct sign; 1 if wrong sign 
    significant_diff = (mae_part > threshold).astype('int')

    # mae in case wrong sign and over threshold
    sign_loss = (np.abs(y_pred) * wrong_signs * significant_diff)
    
    # print("---------------------------------------------------")
    # print(mae_part)
    # print("############")
    # # print(np.sign(y_pred))
    # # print(np.sign(y_true))
    # # print(wrong_signs)
    # # print("############")
    # print(significant_diff)
    # print("############")
    # print(sign_loss)
    # print("---------------------------------------------------")


    return mae_part + sign_loss
```

```python
def sign_loss(y_pred, y_true, threshold = 0.048):
    mae_part = np.abs(y_pred - y_true)
    wrong_signs = (np.sign(y_pred) * np.sign(y_true) == -1).astype('int') # 0 if correct sign; 1 if wrong sign 
    significant_diff = (mae_part > threshold).astype('int')
    sign_loss = (np.abs(y_pred) * wrong_signs * significant_diff)
    return sign_loss
```

```python
mse_by_date = all_predictions.groupby(by=["MODEL", "MONTH"]).apply(lambda x: mse(x["LC(FAT_PRED)"], x["LC(FAT_ACTUAL)"])).reset_index().rename({0: "MSE"}, axis="columns").set_index(["MODEL", "MONTH"], drop=True)
mae_by_date = all_predictions.groupby(by=["MODEL", "MONTH"]).apply(lambda x: meanabs(x["LC(FAT_PRED)"], x["LC(FAT_ACTUAL)"])).reset_index().rename({0: "MAE"}, axis="columns").set_index(["MODEL", "MONTH"], drop=True)
# tadda_by_date = all_predictions.groupby(by=["MODEL", "MONTH"]).apply(lambda x: np.mean(np.abs(x.FAT_PRED-x.FAT_ACTUAL) + sign_loss(x.FAT_PRED, x.FAT_ACTUAL))).reset_index().rename({0: "TADDA"}, axis="columns").set_index(["MODEL", "MONTH"], drop=True)
tadda_by_date = all_predictions.groupby(by=["MODEL", "MONTH"]).apply(lambda x: np.mean(np.abs(x["LC(FAT_PRED)"]-x["LC(FAT_ACTUAL)"]) + sign_loss(x["LC(FAT_PRED)"], x["LC(FAT_ACTUAL)"]))).reset_index().rename({0: "TADDA"}, axis="columns").set_index(["MODEL", "MONTH"], drop=True)

metrics_by_date = pd.concat([mse_by_date, mae_by_date, tadda_by_date], axis=1) # todo: TADDA!
metrics_by_date = metrics_by_date.reset_index()
metrics_by_date
```

```python
def pred_period_plot(substring=""):
    rotate_x_degree = 60
    dep_var = "FATALITIES_LOG_CHANGE"
    metrics = ["TADDA", "MAE", "MSE"]
    # filter models
    selected_models = [string for string in MODELS if substring in string]

    fig, ax = plt.subplots(1, len(metrics), figsize=(20,10))    

    for i, metric in enumerate(metrics):
        plot_df = metrics_by_date[["MONTH", 'MODEL', metric]]
        plot_df = plot_df[plot_df["MODEL"].isin(selected_models)]

        sns.lineplot(x=pd.DatetimeIndex(plot_df['MONTH']), y=plot_df[metric], hue=plot_df['MODEL'], 
                     linewidth=4, ax=ax[i], marker="o")

        ax[i].set_title(dep_var + ' - ' + metric, fontsize=18)
        ax[i].set_ylabel(dep_var + ' - '+ metric, fontsize=18)
        ax[i].set_xlabel('Prediction Period', fontsize=18)

        ax[i].get_legend().remove()
        handles, labels = ax[i].get_legend_handles_labels()
        fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0.05), fontsize=16)
        ax[i].grid()
```
